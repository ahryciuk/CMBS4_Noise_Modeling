# Test Notebook for the Systematics Class

```python
import numpy as np
import os,sys
from fisher_new import fisher
from Systematics import systematics
```

```python
#declare an instance of the systematics class
fsky = 0.4
freq_bands = [95,150]
fwhm_bands = [1.5,1.0]
f3db_beams = [150,150]
ell = np.arange(6,1e4,1.96)

syst = systematics(fsky,freq_bands,fwhm_bands,f3db_beams,ell)
```

```python
import matplotlib.pyplot as plt

for freq in freq_bands:
    plt.plot(ell,beam_unc[freq])
```

# Test 1/f Systematics

```python
#specify user defined inputs for possible 1/f systematics

#readout
NEI_list = [1,1]
S_responsivity = [1e6,2e6]
ell_knee_readout = [200,200]
alpha_ind_readout = [1.5,1.5]

#detector
ell_knee_det = [150,200]
alpha_ind_det = [2.2,2.2]
white_noise_det = [5e-5,3e-5]

#atmosphere
ell_knee_atm = [150,150]
alpha_ind_atm = [2.,2.1]
white_noise_atm = [5e-4,2e-4]

#instantiate dictionaries
readout_spec = {}
det_spec = {}
atm_spec = {}

#generate 1/f spectra for readout, detector, and atmosphere
for freq in freq_bands:
    i = freq_bands.index(freq)
    readout_spec[freq] = syst.oneovereff_readout(ell,NEI_list[i],S_responsivity[i],ell_knee_readout[i],alpha_ind_readout[i])
    det_spec[freq] = syst.oneovereff_det(ell,alpha_ind_det[i],ell_knee_det[i],white_noise_det[i])
    atm_spec[freq] = syst.oneovereff_atm(ell,alpha_ind_atm[i],ell_knee_det[i],white_noise_atm[i])


```

# Plot Spectra

```python
fig,axs = plt.subplots(1,3,figsize=(25,10))

fig.suptitle('1/f Systematics')

for freq in freq_bands:
    #readout plot
    axs[0].plot(ell,readout_spec[freq])
    axs[0].set_title('Readout Systematic Spectrum')
    axs[0].set_xlim(2,1e3)
    axs[0].set_ylim(1e-7,1e-5)

    #detector plot
    axs[1].plot(ell,det_spec[freq])
    axs[1].set_title('Detector Systematic Spectrum')
    axs[1].set_xlim(2,1e3)
    axs[1].set_ylim(1e-7,1e-4)

    #atm plot
    axs[2].plot(ell,atm_spec[freq])
    axs[2].set_title('Atmosphere Systematic Spectrum')
    axs[2].set_xlim(2,1e3)
    axs[2].set_ylim(1e-7,1e-3)
```

# Generate Pseudo Noise Spectrum

```python
#inputs (doesn't include |beam error yet)
white_noise = [5e-4,5e-4]
alpha_ind = [2.2,2.2]
ell_knee = [150,150]

#beam parameters
fwhm_beams = [1.5,1.0]
sigma=[]
for fwhm in fwhm_beams:
    sigma.append(fwhm / np.sqrt(8. * np.log(2)) /60. * np.pi/180.)
    
#generate beam
bl_dic = {}
for freq in freq_bands:
    i = freq_bands.index(freq)
    bl_dic[freq] = np.exp(ell * (ell+1) * sigma[i]**2.)


nl_dic = {}
for freq1 in freq_bands:
    for freq2 in freq_bands:
        i = freq_bands.index(freq1)
        j = freq_bands.index(freq2)
        
        if freq1 == freq2:

            #generate 1/f component
            oneovereff = syst.oneovereff(ell,white_noise[i],ell_knee[i],alpha_ind[i])

            #normalize to white noise level
            nl = np.sqrt(oneovereff**2. + white_noise[i]**2.)*bl_dic[freq1]
        
        else:
            oneovereff_1 = syst.oneovereff(ell,white_noise[i],ell_knee[i],alpha_ind[i])
            oneovereff_2 = syst.oneovereff(ell,white_noise[j],ell_knee[j],alpha_ind[j])
            
            nl = (np.sqrt(oneovereff_1**2. + oneovereff_2**2.) + np.sqrt(white_noise[i]**2. + white_noise[j]**2.)) * \
            np.sqrt(bl_dic[freq1]**2. + bl_dic[freq2]**2.)
            
        #store in dictionary

        nl_dic[(freq1,freq2)] = nl
```

```python
print(nl_dic)
```

# Plot Noise Spectra

```python
for freq1 in freq_bands:
    for freq2 in freq_bands:
        plt.plot(ell,nl_dic[(freq1,freq2)])
        plt.xlim(5,1e4)
        plt.ylim(1e-7,3e-3)
    
```

# Add Systematics to Noise Spectrum

## <font color='orange'>What to do about beam uncertainty? Does weird stuff if just add.</font>

```python

for freq1 in freq_bands:
    for freq2 in freq_bands:
        
        if freq1 == freq2:
            nl_dic[(freq1,freq2)] = np.sqrt(nl_dic[(freq1,freq2)]**2. + readout_spec[freq1]**2. + det_spec[freq1]**2. + \
            atm_spec[freq1]**2.+ bl_dic[freq1]**2.*(beam_unc[freq1])**2.)
        else:
            nl_dic[(freq1,freq2)] = np.sqrt(nl_dic[(freq1,freq2)]**2. + readout_spec[freq1]**2. + readout_spec[freq2]**2. + \
            det_spec[freq1]**2. + det_spec[freq2]**2. + atm_spec[freq1]**2. + atm_spec[freq2]**2. + \
            bl_dic[freq1]**2.*beam_unc[freq1]**2. + bl_dic[freq2]**2.*beam_unc[freq2]**2.)
            
```

## Without Beam Systematic

```python
for freq1 in freq_bands:
    for freq2 in freq_bands:
        
        if freq1 == freq2:
            nl_dic[(freq1,freq2)] = np.sqrt(nl_dic[(freq1,freq2)]**2. + readout_spec[freq1]**2. + det_spec[freq1]**2. + \
            atm_spec[freq1]**2.)
        else:
            nl_dic[(freq1,freq2)] = nl_dic[(freq1,freq2)] + np.sqrt(readout_spec[freq1]**2. + readout_spec[freq2]**2.) + \
            np.sqrt(det_spec[freq1]**2. + det_spec[freq2]**2.) + np.sqrt(atm_spec[freq1]**2.+atm_spec[freq2]**2.)
```

# Plot new Noise Spectra

```python
for freq1 in freq_bands:
    for freq2 in freq_bands:
        plt.plot(ell,nl_dic[(freq1,freq2)])
        plt.xlim(5,1e4)
        plt.ylim(1e-7,3e-3)
```

# Fisher Forecast with nl's

```python
import os,sys

cwd = os.getcwd()
cwd = cwd.rstrip('/ilc_modules')

#inputs (doesn't include beam error yet)
white_noise = [5e-4,5e-4]
alpha_ind = [2.2,2.2]
ell_knee = [150,150]

#beam parameters
fwhm_beams = [1.5,1.0]
sigma=[]
for fwhm in fwhm_beams:
    sigma.append(fwhm / np.sqrt(8. * np.log(2)) /60. * np.pi/180.)
    
#generate beam
bl_dic = {}
for freq in freq_bands:
    i = freq_bands.index(freq)
    bl_dic[freq] = np.exp(ell * (ell+1) * sigma[i]**2.)


nl_dic = {}
for freq1 in freq_bands:
    for freq2 in freq_bands:
        i = freq_bands.index(freq1)
        j = freq_bands.index(freq2)
        
        if freq1 == freq2:

            #generate 1/f component
            oneovereff = syst.oneovereff(ell,white_noise[i],ell_knee[i],alpha_ind[i])

            #normalize to white noise level
            nl = np.sqrt(oneovereff**2. + white_noise[i]**2.)*bl_dic[freq1]
        
        else:
            oneovereff_1 = syst.oneovereff(ell,white_noise[i],ell_knee[i],alpha_ind[i])
            oneovereff_2 = syst.oneovereff(ell,white_noise[j],ell_knee[j],alpha_ind[j])
            
            nl = (np.sqrt(oneovereff_1**2. + oneovereff_2**2.) + np.sqrt(white_noise[i]**2. + white_noise[j]**2.)) * \
            np.sqrt(bl_dic[freq1]**2. + bl_dic[freq2]**2.)
            
        #store in dictionary

        nl_dic[(freq1,freq2)] = nl
```

```python
os.chdir(cwd + '/ilc_modules')
from fisher_new import fisher
os.chdir(cwd)

paramfile = cwd + 'ilc_modules/params_planck_r_0.0_2015_cosmo_lensed_LSS_JM.txt'

which_spectra = 'lensed_scalar'

TParr = ['T','P']
for TP in TParr:
    nl_dic[TP] = nl_dic
TParr = ['TT','EE','TE']
for TP in TParr:
    if TP in ('TT','EE'):
        nl_dic[TP] = nl_dic
    else:
        nl_dic[TP] = None

fisher(paramfile, which_spectra, nl_dic, bl_dic, freq_bands, fwhm_bands, f3db_beams, pct_unc_lpf, \
      ell, fsky, cwd)

```
